```python
Dataframe from Inbuilt Data with Column and Indexes
```

```python
l = [['Alice', 25], ['Bob', 30], ['Charlie', 28]]

list_df = pd.DataFrame(l,columns=['Name','Age'],index=[1,2,3]) 

print(type(list_df)) 
list_df
```

```python
my_tuple = (('Alice', 25), ('Bob', 30), ('Charlie', 28))

tuple_df = pd.DataFrame(my_tuple,columns=['Name','Age'])

tuple_df
```

```python
my_dict = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 29,28]
}

dict_df = pd.DataFrame(my_dict)

dict_df
```

```python
my_data = [
    {'Name': 'Alice', 'Age': 25, 'City': 'New York'},
    {'Name': 'Bob', 'Age': 30, 'City': 'Paris'},
    {'Name': 'Charlie', 'Age': 28}
]

data_df = pd.DataFrame(my_data)
data_df
```
